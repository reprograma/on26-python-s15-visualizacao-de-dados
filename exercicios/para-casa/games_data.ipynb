# Tratamento dos Dados  - Games

```python
import pandas as pd
```

```python
# Leitura do arquivo CSV
df = pd.read_csv("Games.csv")
```

```python
# Verificando informações gerais sobre os dados
print(df.info())
df
```

```python
# Tratamento de valores nulos (substituir [null] por NaN)
df = df.replace('[null]', pd.NA)
df
```

```python
# Removendo linhas com valores nulos
df = df.dropna()
df
```

```python
# Convertendo a coluna 'Sales' para string
df['Sales'] = df['Sales'].astype(str)

# Removendo "M" da coluna 'Sales' e convertendo para float
df['Sales'] = df['Sales'].str.replace('M', '').astype(float)
df
```

```python
# Visualizando as estatísticas descritivas
print(df.describe())
```

```python
# Contagem de valores únicos em cada coluna
unique_counts = df.nunique()
print(unique_counts)
```

```python
# Contagem de valores únicos na coluna 'Publisher'
publisher_counts = df['Publisher'].value_counts()
print(publisher_counts)
```

```python
# Contagem de valores únicos na coluna 'Genre'
genre_counts = df['Genre'].value_counts()
print(genre_counts)
```

```python
# Classificando o DataFrame com base nas vendas (Sales)
df = df.sort_values(by='Sales', ascending=False)
df
```

```python
# Selecionando os 50 jogos mais famosos
top_50_games = df.head(50)
top_50_games
```

```python
# Definindo um dicionário de mapeamento de nomes de colunas em inglês para português
mapeamento_colunas = {
    'Name': 'Nome',
    'Sales': 'Vendas',
    'Series': 'Série',
    'Release': 'Data de Lançamento',
    'Genre': 'Gênero',
    'Developer': 'Desenvolvedor',
    'Publisher': 'Editora'
}

# Renomeando as colunas do DataFrame
top_50_games = top_50_games.rename(columns=mapeamento_colunas)
```

```python
# Nome do novo arquivo CSV
novo_arquivo_csv = "top_50_games.csv"

# Salvando o DataFrame como um novo arquivo CSV
top_50_games.to_csv(novo_arquivo_csv, index=False)
```
