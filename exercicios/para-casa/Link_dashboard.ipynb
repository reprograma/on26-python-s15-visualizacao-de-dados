```python
from IPython.display import HTML, display
import base64

caminho_imagem = r"C:\Users\santi\OneDrive\Imagens\Meu_Painel.png"

def exibir_imagem_base64(caminho):
    with open(caminho, "rb") as image_file:
        encoded_string = base64.b64encode(image_file.read()).decode('utf-8')
    return encoded_string

imagem_base64 = exibir_imagem_base64(caminho_imagem)

html_imagem = f'<img src="data:image/png;base64,{imagem_base64}" alt="Print do painel">'

display(HTML(html_imagem))
```
