# Base utilizada
Foi utilizada uma base de dados do Kaggle que traz informações sobre liberdade e direitos civis entre os países do mundo

```python
import pandas as pd
```

```python
df = pd.read_csv("freedom.csv", sep="," , encoding="latin-1")
df
```

```python
df.info()
```

```python
traducoes = {
    'country': 'pais',
    'year': 'ano',
    'CL': 'civis libertos',
    'PR': 'direitos politicos',
    'status': 'status',
    'Region_Code': 'codigo regiao',
    'Region_Name': 'regiao',
    'is_ldc': 'subdesenvolvido'
}

novas_colunas = []
for chave, valor in traducoes.items():
    novas_colunas.append(valor)

novas_colunas
```

```python
df.columns = novas_colunas
df
```

```python
df.to_csv("base_traduzida.csv", index=False)
```

```python
substituicao = df["subdesenvolvido"].replace([0,1], ["nao", "sim"])
substituicao
```
