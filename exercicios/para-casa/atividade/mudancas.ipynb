## Junção dos CSV

```python
import os

pasta_datasets = '/Users/thaysalima/Desktop/RePrograma./Python Fundamentos e análise de dados/Aula15 11.11/on26-python-s15-visualizacao-de-dados/exercicios/para-casa/paises_escolhidos'


dataframes = []


for arquivo in os.listdir(pasta_datasets):
    if arquivo.endswith('.csv'):
       
        caminho_arquivo = os.path.join(pasta_datasets, arquivo)
        df = pd.read_csv(caminho_arquivo)
        dataframes.append(df)


df_final = pd.concat(dataframes, ignore_index=True)


df_final.to_csv('/Users/thaysalima/Desktop/RePrograma./Python Fundamentos e análise de dados/Aula15 11.11/on26-python-s15-visualizacao-de-dados/exercicios/para-casa/atividade/feriados_2023.csv', index=False)
```
